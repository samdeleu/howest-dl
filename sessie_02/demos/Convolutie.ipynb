```python
%matplotlib inline
import numpy as np
from scipy import ndimage
from scipy import signal
import matplotlib.pyplot as plt
```

```python
# Inlezen image
I = plt.imread('Tesla.png')
```

```python
fig = plt.figure(figsize=(16,9))
plt.imshow(I)
```

```python
# Converteren naar gray image

def rgb2gray(rgb): # Omzetten naar grijswaarden
    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b
    return gray

Ig = rgb2gray(I)
fig = plt.figure(figsize=(10,6))
plt.imshow(Ig,cmap=plt.cm.gray)
```

```python
# kernel
k1 = np.array([[1, 0, -1],
               [2, 0, -2],
               [1, 0, -1]])
#k = np.ones((11,11))*(1/(11**2))
display(k1)
```

```python
# filteren met kernel
If1=signal.convolve2d(Ig,k1)
fig = plt.figure(figsize=(10,6))
plt.imshow(If1,cmap=plt.cm.gray)
```

```python
k2 = np.array([[1, -1, -1],
               [-1, 1, -1],
               [-1, -1, 1]])
If2=signal.convolve2d(Ig,k2)
fig = plt.figure(figsize=(10,6))
plt.imshow(If2,cmap=plt.cm.gray)
```

```python
k3 = np.array([[-1, -1, 1],
               [-1, 1, -1],
               [1, -1, -1]])
If3=signal.convolve2d(Ig,k3)
fig = plt.figure(figsize=(10,6))
plt.imshow(If3,cmap=plt.cm.gray)
```

```python
k4 = np.array([[-1, 1, -11],
               [-1, 1, -1],
               [-1, 1, -1]])
If4=signal.convolve2d(Ig,k4)
fig = plt.figure(figsize=(10,6))
plt.imshow(If4,cmap=plt.cm.gray)
```
